```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import scipy.stats
%matplotlib inline
from sklearn.svm import SVC #SVCを行うためのライブラリ
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_excel('ParkFactor_record.xlsx')
df.head(6)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.set_index(['球場','球団'],inplace=True)
df.head(6)
```

```python
X=scipy.stats.zscore(df)#正規化
df_st=pd.DataFrame(X,columns=df.columns)
df_st.head(6)
```

```python
#平均値PF以上を０、それ以外を１
def over_avg(x):
    if x>=df_st['本塁打PF'].median():
        return 0
    else:
        return 1
df_st['over_avg']=df_st['本塁打PF'].apply(over_avg)
df_st.head(6)
```

```python
X2=df_st.drop(['本塁打PF','over_avg'],axis=1)#説明変数
Y2=df_st['over_avg']#目的変数
```

```python
#テストデータと訓練データを分ける
X2_train,X2_test,Y2_train,Y2_test=train_test_split(X2,Y2,test_size=0.2,shuffle=True)
```

```python
#SVMモデルを準備
model=SVC(C=1.0,kernel='linear')
#学習する
model.fit(X2_train,Y2_train)
```

```python
#関係性をデータフレームに格納
model_coeff=DataFrame([X2.columns,model.coef_[0]]).T
#実行
model_coeff
```
