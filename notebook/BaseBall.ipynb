```python
df_hr=df[df["本塁打"]>20]#本塁打が２０より大きい行を抽出
df_hr.head()
```

```python
df['盗塁成功率']=df['盗塁']/(df['盗塁']+df['盗塁死'])#盗塁成功率の追加
df.head(6)
```

```python
df.rename(columns={'年度':'年'},inplace=True)#列名を変更
df.head(6)
```

```python
df.rename(columns={'年':'年度'},inplace=True)
df.head(6)
```

```python
df.sort_values(by='安打',ascending=False).head()#安打が多い順に並び替え、False＝昇順
```

```python
df.sort_values(by='年度',ascending=True).head()#年度が少ない順に並び替え、True＝降順
```

```python
df2=pd.read_excel('Ichiro2.xlsx',na_values='?')
df2.head(3)
```

```python
df3=pd.concat([df,df2])#Ichiro2.xlsxとIchiro.xlsxを結合したdf3を作る
df3.tail(6)
```

```python
df.isnull().sum()#ファイルの欠損を確認
```

```python
df.fillna(value={'盗塁成功率':0},inplace=True)#データの欠損部分を０に
df.tail(5)
```

```python
df.dropna(subset=['盗塁成功率'],axis=0,inplace=True)#欠損のある行を削除
pd.DataFrame(df['盗塁成功率'].mask(df['盗塁成功率']=NaN,0)).head()#盗塁成功率がNaNの場合、０の置換する
```
